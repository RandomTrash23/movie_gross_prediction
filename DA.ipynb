```python
df.dropna(subset=['budget'],inplace=True)
df.info()
```

```python
df.dropna(subset=['gross'],inplace=True)
df.info()
```

```python
df.describe()
```

```python
df.drop(labels='released',axis=1,inplace=True)
df.info()
```

```python
min_gross=df['gross'].idxmin()
df.iloc[min_gross]
```

```python
df['profit_percent']=((df.gross-df.budget)/(df.budget))*100
```

```python
df.groupby('genre')['profit_percent'].agg(['count','min','max','mean','median']).sort_values(by='mean',ascending=False)
```

```python
d1 = df.groupby('genre')['profit_percent'].agg(['count','median']).sort_values(by='count',ascending=False)
d1 = d1.reset_index()
df_melted = d1.melt(id_vars='genre', value_vars=['count', 'median'], var_name='Statistic', value_name='Value')
plt.figure(figsize=(12, 8))
sns.barplot(x='Value', y='genre', hue='Statistic', data=df_melted, palette='viridis')
plt.title('Count, and Median Profit Percentage by Genre')
plt.xlabel('Value')
plt.ylabel('Genre')
plt.legend(title='Statistic')
plt.show()
```

```python
df.groupby('rating')['profit_percent'].agg(['count','min','max','mean','median']).sort_values(by='mean',ascending=False)
```

```python
d2 = df.groupby('rating')['profit_percent'].agg(['count','median']).sort_values(by='count',ascending=False)
d2 = d2.reset_index()
df_melted_2 = d2.melt(id_vars='rating', value_vars=['count', 'median'], var_name='Statistic', value_name='Value')
plt.figure(figsize=(12, 8))
sns.barplot(x='Value', y='rating', hue='Statistic', data=df_melted_2, palette='viridis')
plt.title('Count, and Median Profit Percentage by rating')
plt.xlabel('Value')
plt.ylabel('rating')
plt.legend(title='Statistic')
plt.show()
```

```python
df.groupby('genre')['score'].agg(['count','min','max','mean','median']).sort_values(by='mean',ascending=False)
```

```python
d3 = df.groupby('genre')['score'].agg(['median','mean']).sort_values(by='median',ascending=False)
d3 = d3.reset_index()
df_melted_3 = d3.melt(id_vars='genre', value_vars=['median','mean'], var_name='Statistic', value_name='Value')
plt.figure(figsize=(12, 8))
sns.barplot(x='Value', y='genre', hue='Statistic', data=df_melted_3, palette='viridis')
plt.title('median and mean score by genre')
plt.xlabel('Value')
plt.ylabel('Genre')
plt.legend(title='Statistic')
plt.show()
```
